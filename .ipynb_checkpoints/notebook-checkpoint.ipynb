## Topic Clustering

```python
 >>> import nltk
 >>> nltk.download('wordnet')
```

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *
import numpy as np

##Definimos unas funciones para el preprocesamiento del texto
#Hacemos una prueba para ver que este funcionando


def lemmatize_stemming(text):
    return SnowballStemmer('english').stem(WordNetLemmatizer().lemmatize(text, pos='v'))
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(str(text)):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return result
print(df_new.title.values[0])
preprocess(df_new.title.values[0])
```

```python
df_new['processed_docs'] = list(map(preprocess,df_new['description'].values))
df_new.head(5)
```

```python
#Aplicamos el preprocesamiento a la columna de interes
processed_docs = list(map(preprocess,df_new['description'].values))
#Creamos un diccionario de palabras con gensim
dictionary = gensim.corpora.Dictionary(processed_docs)
dictionary.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
#Creamos nuestro bog generado por gensim de cada articulo
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
bow_corpus[4310]
```

```python
bow_doc_4310 = bow_corpus[4310]
# Una manera mas practica de visualizar de visualizar nuestro BOW
for i in range(len(bow_doc_4310)):
    print("Word {} (\"{}\") appears {} time.".format(bow_doc_4310[i][0], 
                                               dictionary[bow_doc_4310[i][0]], 
bow_doc_4310[i][1]))
```

```python
# Una vez ya tenemos preparado nuestro corpus y nuestro diccionario, se pueden ingresar estos al modelo de LDA de gensim y ver que resultados nos entrega

lda_model = gensim.models.LdaMulticore(bow_corpus, num_topics=10, id2word=dictionary, passes=2, workers=2)
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
    #Este ciclo for nor permite vizualizar los topicos o clusters, generados segun la data ingresada.
    #Ademas retorna las palabras con mayor peso que definen el topico en cuestion
```

```python
# Creamos un empaquetado de lo anterior para que sea mas facil el generar estos modelos de topicos

class topic_generator:
    def lemmatize_stemming(text):
        return SnowballStemmer('english').stem(WordNetLemmatizer().lemmatize(text, pos='v'))
    def preprocess(text):
        result = []
        for token in gensim.utils.simple_preprocess(str(text)):
            if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
                result.append(lemmatize_stemming(token))
        return result
    
    
    def __init__(self,list_text,num_topics=10,passes=2,workers=4):
        #Se debe ingresar una lista con los documentos a procesar
        self.text=list_text
        processed_docs = list(map(preprocess,list_text))
        self.dictionary = gensim.corpora.Dictionary(processed_docs)
        self.dictionary.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
        bow_corpus = [self.dictionary.doc2bow(doc) for doc in processed_docs]
        self.lda_model={}
        self.lda_model['bow']= gensim.models.LdaMulticore(bow_corpus, num_topics=num_topics, id2word=self.dictionary, passes=passes, workers=workers)
        self.tfidf = gensim.models.TfidfModel(bow_corpus)
        corpus_tfidf = self.tfidf[bow_corpus]
        self.lda_model['tfid'] = gensim.models.LdaMulticore(corpus_tfidf, num_topics=num_topics, id2word=self.dictionary, passes=passes, workers=workers)
        
    def __str__(self):
        texto="Topics modelo con Bag of words: "
        for idx, topic in self.lda_model['bow'].print_topics(-1):
            texto+='\nTopic: {} \nWords: {}'.format(idx, topic)
        texto+="\n\nTopics modelo con Tfid: "
        for idx, topic in self.lda_model['tfid'].print_topics(-1):
            texto+='\nTopic: {} \nWords: {}'.format(idx, topic)
        return texto
            
    def predict_topic(self,corpus,model=None):
        if (model is None) | (model not in ["bow","tfid"]):
            print("Se debe ingresar el modelo con el cual se realizara la prediccion(bow, tfid)")
        else:
            topic=[]
            for corpu in corpus:
                topic.append(sorted(self.lda_model[model][corpu], key=lambda tup: -1*tup[1])[0][0])
            return topic
            
    
    def preprocess_text(self,list_text,model=None):
        if (model is None) | (model not in ["bow","tfid"]):
            print("Se debe ingresar el modelo con el cual se realizara la prediccion(bow, tfid)")
        else:
            processed_docs = list(map(preprocess,list_text))
            bow_corpus = [self.dictionary.doc2bow(doc) for doc in processed_docs]
            if model == "bow":
                return bow_corpus
            else:
                return self.tfidf[bow_corpus]
    
```

```python
Topic_model=topic_generator(df_new['title'].values)
```

```python
corpus=Topic_model.preprocess_text(df_new['title'].values[0:3],model='bow')
Topic_model.predict_topic(corpus,model='bow')
```

Se creo un modelo funcional que permite crear de manera rapida y sencilla un clasificador por topicos.

Observando nuestros resultados, esta claro que se necesitan un dataset mucho mas grande y con noticias mas variadas, que permita que el modelo cree unos clusters de topicos mucho mas precisos, ya que por ahora se mira que la mayoria de los topicos que se generan estan relacionados a temas de politicos de USA
