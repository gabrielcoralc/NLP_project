```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *
import numpy as np

##Definimos unas funciones para el preprocesamiento del texto
#Hacemos una prueba para ver que este funcionando


def lemmatize_stemming(text):
    return SnowballStemmer('english').stem(WordNetLemmatizer().lemmatize(text, pos='v'))
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(str(text)):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return result
print(df_new.title.values[0])
preprocess(df_new.title.values[0])
```

```python
df_new['processed_docs'] = list(map(preprocess,df_new['description'].values))
df_new.head(5)
```

```python
#Aplicamos el preprocesamiento a la columna de interes
processed_docs = list(map(preprocess,df_new['description'].values))
#Creamos un diccionario de palabras con gensim
dictionary = gensim.corpora.Dictionary(processed_docs)
dictionary.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
#Creamos nuestro bog generado por gensim de cada articulo
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
bow_corpus[4310]
```

```python
# Creamos un empaquetado de lo anterior para que sea mas facil el generar estos modelos de topicos

class topic_generator:
    def lemmatize_stemming(text):
        return SnowballStemmer('english').stem(WordNetLemmatizer().lemmatize(text, pos='v'))
    def preprocess(text):
        result = []
        for token in gensim.utils.simple_preprocess(str(text)):
            if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
                result.append(lemmatize_stemming(token))
        return result
    
    
    def __init__(self,list_text,num_topics=10,passes=2,workers=4):
        #Se debe ingresar una lista con los documentos a procesar
        self.text=list_text
        processed_docs = list(map(preprocess,list_text))
        self.dictionary = gensim.corpora.Dictionary(processed_docs)
        self.dictionary.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
        bow_corpus = [self.dictionary.doc2bow(doc) for doc in processed_docs]
        self.lda_model={}
        self.lda_model['bow']= gensim.models.LdaMulticore(bow_corpus, num_topics=num_topics, id2word=self.dictionary, passes=passes, workers=workers)
        self.tfidf = gensim.models.TfidfModel(bow_corpus)
        corpus_tfidf = self.tfidf[bow_corpus]
        self.lda_model['tfid'] = gensim.models.LdaMulticore(corpus_tfidf, num_topics=num_topics, id2word=self.dictionary, passes=passes, workers=workers)
        
    def __str__(self):
        texto="Topics modelo con Bag of words: "
        for idx, topic in self.lda_model['bow'].print_topics(-1):
            texto+='\nTopic: {} \nWords: {}'.format(idx, topic)
        texto+="\n\nTopics modelo con Tfid: "
        for idx, topic in self.lda_model['tfid'].print_topics(-1):
            texto+='\nTopic: {} \nWords: {}'.format(idx, topic)
        return texto
            
    def predict_topic(self,corpus,model=None):
        if (model is None) | (model not in ["bow","tfid"]):
            print("Se debe ingresar el modelo con el cual se realizara la prediccion(bow, tfid)")
        else:
            topic=[]
            for corpu in corpus:
                topic.append(sorted(self.lda_model[model][corpu], key=lambda tup: -1*tup[1])[0][0])
            return topic
            
    
    def preprocess_text(self,list_text,model=None):
        if (model is None) | (model not in ["bow","tfid"]):
            print("Se debe ingresar el modelo con el cual se realizara la prediccion(bow, tfid)")
        else:
            processed_docs = list(map(preprocess,list_text))
            bow_corpus = [self.dictionary.doc2bow(doc) for doc in processed_docs]
            if model == "bow":
                return bow_corpus
            else:
                return self.tfidf[bow_corpus]
    
```

## Title Generation

```python
# keras module for building LSTM 
from keras.preprocessing.sequence import pad_sequences
from keras.layers import Embedding, LSTM, Dense, Dropout
from keras.preprocessing.text import Tokenizer
from keras.callbacks import EarlyStopping
from keras.models import Sequential
import keras.utils as ku 
import tensorflow as tf
# set seeds for reproducability

from numpy.random import seed

seed(1)

import string, os 

import warnings
warnings.filterwarnings("ignore")
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Creamos otra funcion para limpiar el texto, similiar a la anterior pero en este caso aun no necesitamos separar las palabras ni quitar las
#stopwords, dado que queremos generar titulos similares a los ya existentes



def preprocess_v2(text):
    result = ""
    for token in gensim.utils.simple_preprocess(str(text)):
            result+=" "+token
    return result.strip()

df['clean_title']=list(map(preprocess_v2,df.title.values))
df.head(5)
```

```python
# Creamos los tokens de manera que puedan ser ingresados al modelo de LTSM.
#Ingresamos las cadenas de texto de tal manera que vamos agregando una palabra a la vez
#De esta manera el modelo tendra que aprender cual es la palabra que deberia seguir segun la entrada que este reciba


def get_sequence_of_tokens(corpus):
    
    ## tokenization
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(corpus)
    total_words = len(tokenizer.word_index) + 1
    
    ## convert data to sequence of tokens 
    input_sequences = []
    for line in corpus:
        token_list = tokenizer.texts_to_sequences([line])[0]
        for i in range(1, len(token_list)):
            n_gram_sequence = token_list[:i+1]
            input_sequences.append(n_gram_sequence)
    return input_sequences, total_words,tokenizer

inp_sequences, total_words,tokenizer = get_sequence_of_tokens(df.clean_title.values[:1000])
inp_sequences[:10]
```

```python
#Padding the Sequences and obtain Variables : Predictors and Target

def generate_padded_sequences(input_sequences,total_words):
    max_sequence_len = max([len(x) for x in input_sequences])
    input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')) #Esta funcion permite que todas las cadenas
                                                                                                       # De texto tenga el mismo largo
    
    predictors, label = input_sequences[:,:-1],input_sequences[:,-1] #Separamos la ultima palabra de cada cadena con el fin de que este sea nuestro valor
                                                                     #A predecir
    label = tf.keras.utils.to_categorical(label, num_classes=total_words) #Esta funcion nos devuelve una cadena de 0's, con un 1 en la ubicacion de la palabra
                                                              # segun el corpus que realizamos anteriormente
    return predictors, label, max_sequence_len

predictors, label, max_sequence_len = generate_padded_sequences(inp_sequences,total_words)
```

```python
#Creacion de la red neuronal

def create_model(max_sequence_len, total_words):
    input_len = max_sequence_len - 1
    model = Sequential()
    
    # Add Input Embedding Layer
    model.add(Embedding(total_words, 10, input_length=input_len)) #La entrada del modelo la cual lo reducimos todo a 10 neuronas
    
    # Add Hidden Layer 1 - LSTM Layer
    model.add(LSTM(100)) # La capa LTSM la principal encargada de aprender que palabra deberia ser la siguiente
    model.add(Dropout(0.1))
    
    # Add Output Layer
    model.add(Dense(total_words, activation='softmax')) #se crea la salida con un numero de neuronas igual al numero de palabras que tiene nuestro corpus

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    return model

model = create_model(max_sequence_len, total_words)
model.summary()
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint

# agregamos unos callbacks que nos permiten monitorear nuestro modelo

es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=30)
mc = ModelCheckpoint('best_model.h5', monitor='val_accuracy', mode='max', verbose=1, save_best_only=True)
history=model.fit(predictors, label, epochs=100, verbose=1,callbacks=[es,mc],validation_split=0.33)
```

```python
# Observamos la grafica de aprendizaje

import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

Como se puede observar los resultados no son exactamente los mejores, pero este es un modelo de practica para demostrar la posibilidad de enseñarle a una red neuroal el como generar texto a partir de unas posibles soluciones.

Estos modelos de generacion de texto requieren mucha informacion, ademas de que requieren bastante capacidad computacional, por lo que solo hicimos 100 epocas de entrenamiento para ver los resultados que un primer modelo puede brindar.

```python
##Ya con nuestro modelo entrenado, necesitamos una manera practica de empezar a generar texto para nuestro nuevos titulares

def generate_text(seed_text, next_words,tokenizer, model, max_sequence_len):
    for _ in range(next_words):
        # Realizamos el mismo preprocesamiento del texto para que este quede en el mismo formato con el que se entreno el modelo
        token_list = tokenizer.texts_to_sequences([seed_text])[0] 
        token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        predicted = np.argmax(model.predict(token_list, verbose=0),axis=-1)
        # No olvidemos que nuestro modelo nos regresa un numero, que indica la posicion de la palabra en nuestro corpus
        output_word = ""
        for word,index in tokenizer.word_index.items(): # identificamos la palabra que corresponde al indice y lo retornamos
            if index == predicted:
                output_word = word
                break
        seed_text += " "+output_word
    return seed_text.title()
generate_text("NTSB says Autopilot", 10,tokenizer, model, max_sequence_len)
```

Como vemos el modelo funciona, aunque aun no sabe cual deberia ser el largo adecuado de este titular para que este genere una oracion con coherencia.

Como se menciono anteriormente este tipo de modelos requiere de mas parametros(palabras) y un mayor tiempo de entrenamiento, para poder generar resultados mas precisos.

```python
## Para la generacion de titulos tambien podemos probar con los n-grams mas representativos de las descripcion o del contenido
# de los articulos

from sklearn.feature_extraction.text import TfidfVectorizer

df['clean_description']=list(map(preprocess_v2,df.description.values))
vectorizer = TfidfVectorizer(ngram_range=(4, 4)) # Aqui definimos el rango de los n-gramas con los que se va a trabajar

vectorizer.fit(df.clean_description.values)
```

```python
#Creamos una funcion que nos permita extraer los ngramas mas caracteristicos de manera mucho mas rapida

def get_relevant_ngrams(vectorizer,text_list,num_ngrams):
    # vectorizer =  objeto TfidfVectorizer entrenado previamente con el texto a analizar
    # text_list = una lista que contiene la string a procesar
    # num_ngrams= el numero de ngrams que se quiere obtener
    tfid_text=vectorizer.transform(text_list).toarray()[0]
    ngram_importance=sorted(tfid_text)
    ngram_index=[]
    index=0
    while len(ngram_index)<num_ngrams:
        index=index-1-len(ngram_index)
        indices=np.where(tfid_text == ngram_importance[index])[0]
        if len(ngram_index)+len(indices)<=num_ngrams:
            ngram_index=np.append(ngram_index,indices)
        else:
            ngram_index=np.append(ngram_index,indices[:num_ngrams-(len(ngram_index)+len(indices))])
    return list(map(lambda x: vectorizer.get_feature_names()[int(x)],ngram_index))

get_relevant_ngrams(vectorizer,df.clean_description.values[0:1],10)
```
